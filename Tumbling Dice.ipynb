```python
#Example 4 Vol1
import random
from gtts import gTTS
import os
import pyttsx3
from pygame import mixer
import playsound
c1 = 0 
c2 = 0 
c3 = 0 
c4 = 0
c5 = 0
c6 = 0 
p1 = 0
p2 = 0 
p3 = 0 
p4 = 0 
p5 = 0 
p6 = 0 
N = 0

with open("RollingDice.txt" ,"w")as f1:
    content = f1.write("Hello There , Please Enter The Number Of The Throws ")
with open("RollingDice.txt" , "r") as f1 :
    content = f1.read()
    
#Initialize The Properties Of Audio Text   
text = pyttsx3.init("sapi5")
voice = text.getProperty('voices')
text.setProperty('voice',voice[1].id)

rate = text.getProperty("rate")
text.setProperty("rate" , 150)#150 is Normal Speed Of The Audio Message

text.say(content)
text.runAndWait()

N = int(input("Input Numbers:"))    
dice = random.randint(1 , 100)
print("Start Rolling") 
for dice in range (N):
    if dice >=0 and dice <= 10 :
        c1 += 1
    if dice > 10 and dice <= 30:
        c2 += 1
    if dice > 30 and dice <= 50:
        c3 += 1
    if(dice > 50 and dice <= 60):
        c4 += 1
    if(dice > 60 and dice <= 65):
        c5 += 1
    if (dice < 65 and dice > 90):
        c6 += 1
    
        
#p=[p1,p2,p3,p4,p5,p6]
    p1 = c1 / N
    p2 = c2 / N
    p3 = c3 / N
    p4 = c4 / N
    p5 = c5 / N
    p6 = c6 / N
    
    playsound.playsound("Dice1.mp3")#Everytime We Roll The Dice We Hear The Sound
    print(" 1 is : %f" %p1)
    print(" 2 is : %f" % p2)
    print(" 3 is : %f" %p3 )
    print(" 4 is : %f" %p4)
    print(" 5 is : %f" %p5)
    print(" 6 is : %f" %p6)
    print("Stop Rolling")
```
